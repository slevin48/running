Analyze GPX data with Python & MATLAB
https://pypi.org/project/gpxpy/

```python
import gpxpy
import gpxpy.gpx
```

```python
import os
gpx_files = os.listdir('tomtom')
```

```python
gpx_files[-1]
```

```python
gpx_file = open('tomtom/'+gpx_files[-1])
```

```python
gpx = gpxpy.parse(gpx_file)
```

```python
track = gpx.tracks[0]
```

```python
Seg = track.segments
Seg
```

```python
Run0 = [point for point in track.segments[0].points]
Run0[0:5]
```

```python
Run = [point for segment in track.segments for point in segment.points]
```

```python
Lat = [point.latitude for point in Run]
Lon = [point.longitude for point in Run]
```

```python
route = [[x,y] for x in Lat for y in Lon]
route[0:5]
```

```python
for track in gpx.tracks:
    for segment in track.segments:
        for point in segment.points:
            print('Point at ({0},{1}) -> {2}'.format(point.latitude, point.longitude, point.elevation))
```

```python
for waypoint in gpx.waypoints:
    print('waypoint {0} -> ({1},{2})'.format(waypoint.name, waypoint.latitude, waypoint.longitude))
```

```python
for route in gpx.routes:
    print('Route:')
    for point in route.points:
        print('Point at ({0},{1}) -> {2}'.format(point.latitude, point.longitude, point.elevation))
```
